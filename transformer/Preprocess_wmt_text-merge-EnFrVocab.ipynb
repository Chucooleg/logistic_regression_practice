### Manually trim non-latin characters

```python
en_inv_vocab[26498-2752]
```

```python
fr_vocab_trimed = {v:int(k) for k,v in fr_inv_vocab.items() if int(k) < (31267-2199)}
```

### Merge english and french vocab

```python
# total vocab size
print(len(set(fr_vocab_trimed.keys()) | set(en_vocab_trimed.keys())))
```

```python
merged_set = list((set(fr_vocab_trimed.keys()) | set(en_vocab_trimed.keys())) - set(['<EOS>', '<GO>', '<UNK>', '']))
```

```python
merged_vocab = {'<GO>':0, '<EOS>':1, '<PAD>':2, '<UNK>':3}

for typ in merged_set:
    merged_vocab[typ] = len(merged_vocab)
```

```python
merged_inv_vocab = {v:k for k,v in merged_vocab.items()}
```

## 3. Preprocess corpus

```python
def pp_corpus(corpus_lines, word2idx_lookup):
    symbol_corpus = []
    for ln in corpus_lines:
        tokens = ln.rstrip('\n').split(' ')
        symbols = [word2idx_lookup.get(tok, word2idx_lookup['<UNK>']) for tok in tokens]
        symbol_corpus.append(symbols)
    return symbol_corpus
```

```python
en_pp = pp_corpus(corpus_lines=en, word2idx_lookup=merged_vocab)
fr_pp = pp_corpus(corpus_lines=fr, word2idx_lookup=merged_vocab)
```

```python
with open('/home/legg/wmt-2014/training/concat.fr-en.BPE_pp.en', 'w') as f:
    json.dump(en_pp, f)
    
with open('/home/legg/wmt-2014/training/concat.fr-en.BPE_pp.fr', 'w') as f:
    json.dump(fr_pp, f)
```

```python
en_dev_pp = pp_corpus(corpus_lines=en_dev, word2idx_lookup=merged_vocab)
fr_dev_pp = pp_corpus(corpus_lines=fr_dev, word2idx_lookup=merged_vocab)
```
